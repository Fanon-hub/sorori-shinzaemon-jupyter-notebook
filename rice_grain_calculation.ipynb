```python
import matplotlib.pyplot as plt 

days =100 # parameters
daily_rice =[2**(day-1) for day in range(1,days+1)] #variables for calculations
    
# rice received each day 
cumulative_rice =[sum(daily_rice[:day]) for day in range(1, days+1)] #cumulative rice
    
    # total rice recived on the 100th day
total_rice_100_days =cumulative_rice[-1]
    
    # Output
print(f"Total grains of rice received up to the 100th day:{total_rice_100_days}")
plt.figure(figsize=(12,6))
plt.plot(range(1, days+1), daily_rice, label="Daily Rice received", color="yellow")
plt.plot(range(1, days+1), cumulative_rice, label="Cumulative Rice Received", color="green")
    
plt.yscale("log")
plt.title("Grains of rice received over 100 Days")
plt.xlabel("Number of Days")
plt.ylabel("Number of Grains of Rice (Log Scale)")
plt.legend()
plt.grid(True, which="both", linestyle="-.", linewidth=0.5)
    
    
plt.show()
```
